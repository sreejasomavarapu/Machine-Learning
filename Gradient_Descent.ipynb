```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Salary_Data.csv')
```

```python
x = dataset.iloc[:,0]
y = dataset.iloc[:,1]
```

```python
plt.scatter(x,y)
```

```python
m=0
c=0

L=0.0001
iter = 1000
n=float(len(x))
for i in range(iter):
  y_pred = m*x + c
  dm = -2*(sum((y-y_pred)*x))/n
  dc = -2*(sum(y-y_pred))/n
  m = m - L*dm
  c = c - L*dc
y_pred = m*x + c 
```

```python
plt.scatter(x,y)
plt.plot(x,y_pred,color='red')
plt.show()
```
