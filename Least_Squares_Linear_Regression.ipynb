```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('Salary_Data.csv')
x=dataset.iloc[:,0]
y=dataset.iloc[:,1]
plt.scatter(x,y)
plt.show()
```

```python
xm = np.mean(x)
ym = np.mean(y)
num=0
den=0
for i in range(len(x)):
  num += (x[i]-xm)*(y[i]-ym)
  den += (x[i]-xm)**2
m = num/den
c = ym - m*xm

```

```python
y_pred = m*x + c
plt.scatter(x,y)
plt.plot(x,y_pred,color='red')
```
